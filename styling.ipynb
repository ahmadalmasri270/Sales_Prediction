```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

load data set

```python
df=pd.read_csv('/content/drive/MyDrive/Coding Dojo/athleteEventsNoPersonal.csv')
df.tail()
```

## get number of olympians per year

```python
from pandas.core.groupby.groupby import GroupBy
num_uniq_year= df.groupby(['Year'])['ID'].nunique()
print(num_uniq_year.max())
# this is a pandas Series
num_uniq_year
```

```python
# Can convert to NumPy array or list
num_uniq_yearin = np.array(num_uniq_year.index)
#numOlympian = np.array(num_uniq_year.values)
num_uniq_yearin
```

```python
numOlympian = np.array(num_uniq_year.values)
numOlympian
```

```python
# Make the style by default seaborn
plt.style.use('seaborn')
```

```python
# Plot graph showing how it would look with defaults
plt.plot(num_uniq_yearin, numOlympian, c= 'k')
```

```python
#MATLAB-style
# Set xlim and ylim
plt.plot(num_uniq_yearin, numOlympian, c = 'magenta', marker = '*', markersize = 10)
plt.xlim(left=1892,right=2020)
plt.ylim(bottom=0,top=1800)
plt.xlabel('Year')
plt.ylabel('Number of Olympians')
plt.title('Number of Olympians Over the Years')
```

```python
# Change fontsize
fig, axes = plt.subplots(nrows = 1, ncols = 1)
axes.plot(num_uniq_yearin, numOlympian, c= 'k');
axes.set_xlim(left=1892,right=2020)
axes.set_ylim(bottom=0,top=1800)
axes.set_xlabel('Year', fontsize = 16)
axes.set_ylabel('Number of Olympians', fontsize = 16)
axes.set_title('Number of Olympians Over the Years', fontsize = 16)
```
